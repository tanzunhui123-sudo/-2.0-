```python
# Test specific models
test_ids = [
    'gemini-3.1-pro',
    'gemini-3-pro-preview',
    'gemini-3-pro-image-preview',
    'gemini-2.5-pro-preview-05-06',
    'gemini-2.0-flash',
]

for mid in test_ids:
    try:
        model = genai.GenerativeModel(mid)
        resp = model.generate_content('Say hi in one word')
        print(f'  ✅ {mid}: {resp.text.strip()[:60]}')
    except Exception as e:
        print(f'  ❌ {mid}: {str(e)[:80]}')
```
